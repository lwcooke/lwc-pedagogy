# Fourier Series

Here's a quick demonstration of Fourier series; how we can represent functions with them, and ways we can use them to help understand some properties of these functions (with special application to physics).

```julia
using Plots, LaTeXStrings, FFTW
```

```julia
function sinwave(f, t)
    """ Returns sinusoid with frequency f at time t. """

    return sin(2Ï€ * f * t)
end
```

```julia
time = 0:1/1000:1  # Time axis [s]

plot()
for freq in 1:3
    plot!(time, sinwave.(freq, time), label=L"%$(freq)\ \mathrm{Hz}") 
end
xlabel!("Time [s]")
ylabel!("Amplitude [arb]")
```
